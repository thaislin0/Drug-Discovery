```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colors as mcolors

```

```python
url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression = 'zip')
dados.head()
```

```python
plt.bar('Com droga', dados['tratamento'].value_counts(normalize=True)[0], color = 'darkorange', align='center', data=None, width=0.7)
plt.bar('Com tratamento', dados['tratamento'].value_counts(normalize=True)[1], color = 'gold', align='center', data=None, width=0.7)
plt.xlabel("Tipo de tratamento", color = 'peru')
plt.ylabel("Porcentagem", color='peru')
plt.title("Análise de Tratamentos", color='peru')
```

```python
labels = '2 dias (48 horas)', '3 dias (72 horas)', '1 dia (24 Horas)'
sizes = dados['tempo'].value_counts(normalize=True)
explode = (0.1, 0, 0)
```

```python
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal') 
plt.suptitle('Tempo da dosagem', color='peru')
```

```python
plt.bar('Dose D1', dados['dose'].value_counts(normalize=True)[0], color = 'darkcyan', align='center', data=None, width=0.5)
plt.bar('Dose D2', dados['dose'].value_counts(normalize=True)[1], color = 'c', align='center', data=None, width=0.5)
plt.xlabel("Tipo de tratamento", color = 'peru')
plt.ylabel("Quantidade", color='peru')
plt.title("Tipos de dose", color='peru')
```
